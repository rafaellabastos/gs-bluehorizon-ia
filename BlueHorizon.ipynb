## **Detecção e classificação de lixos com deep learning**

<p>Utilização de técnicas de identificação e classificação de lixos em praias e oceanos, sejam imagens subaquáticas ou superficiais, com o objetivo de desenvolver um modelo que possa distinguir entre diferentes tipos de lixo para que ajude a detectar o lixo coletado nos pontos de coleta da BlueHorizon.</p>

<p>Tendo em vista que o objetivo da detecção e classificação dos lixos do nosso projeto será para ajudar na coleta do lixo, os testes foram alimentados com imagens de diferentes formas que as pessoas podem colocar na esteira para a IA ler.</p>

<p>Além disso, como o voluntário pode colocar dois tipos de lixo junto para a detecção (o que não pode), a IA também foi treinada para detectar se isso acontecer.</p>

```python
# Importação dos pacotes
from roboflow import Roboflow
import cv2
from google.colab.patches import cv2_imshow
```

```python
# Acesso ao projeto do Roboflow pela chave de api e iniciação dele
rf = Roboflow(api_key="EC09vGcQVUOr44aasZmU")
project = rf.workspace("projeto-l2c3r").project("bluehorizon-xyzfs")
version = project.version(2).model
```

### **Identificação e classificação dos lixos**

```python
# Imagem de uma lata junto com isopor
version.predict("LataIsopor.jpg", confidence=5, overlap=60).save("LATAISOPOR.jpg")
```

```python
# Resultado da detecção
resultado = cv2.imread('LATAISOPOR.jpg')
cv2_imshow(resultado)
```

```python
# Imagem de um cigarro
version.predict("Cigarro.jpg", confidence=1, overlap=60).save("CIGARRO.jpg")
```

```python
# Resultado da detecção
resultado = cv2.imread('CIGARRO.jpg')
cv2_imshow(resultado)
```

```python
# Imagem de vários cigarros
version.predict("Cigarros.jpg", confidence=20, overlap=60).save("CIGARROS.jpg")
```

```python
# Resultado da detecção
resultado = cv2.imread('CIGARROS.jpg')
cv2_imshow(resultado)
```

```python
# Resultado da detecção
resultado = cv2.imread('SACOLA.jpg')
cv2_imshow(resultado)
```
